```python
# 2. numpy array attributes
arr1=np.array([[1,2,3],[4,5,6]])
print(arr1)
```

```python
print("shape:",arr1.shape)#says number of rows and column
print("size:",arr1.size)#says how much elements in the array
print("dtype:",arr1.dtype) #says the datatype of the elements inside the array
print("ndim:",arr1.ndim)#says the dimension of the array
```
